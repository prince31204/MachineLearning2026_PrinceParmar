```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===============================
# 🎨 GLOBAL COLOR & STYLE SETTINGS
# ===============================

# Seaborn theme (very colorful & modern)
sns.set_theme(
    style="darkgrid",
    palette="viridis",   # color palettes: viridis, plasma, magma, coolwarm, Set2
    font_scale=1.2
)

# Matplotlib settings
plt.rcParams['figure.figsize'] = (12, 7)
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# ===============================
# 📂 LOAD DATASET
# ===============================

medical_df = pd.read_csv(
    r"D:\machinelearning\MachineLearning2026_PrinceParmar\Datasets\medical.csv"
)

print("Dataset loaded successfully ✅")
display(medical_df.head())

# ===============================
# 📊 1. HISTOGRAM (COLORFUL)
# ===============================

sns.histplot(
    medical_df['age'],
    kde=True,
    bins=30,
    color="orange"
)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# ===============================
# 📦 2. BOXPLOT (MULTI COLOR)
# ===============================

sns.boxplot(
    x="sex",
    y="charges",
    data=medical_df,
    palette="Set2"
)

plt.title("Medical Charges by Gender")
plt.show()

# ===============================
# 🔵 3. SCATTER PLOT (VIVID)
# ===============================

sns.scatterplot(
    x="age",
    y="charges",
    hue="smoker",
    data=medical_df,
    palette="coolwarm",
    s=100
)

plt.title("Age vs Charges (Smoker vs Non-Smoker)")
plt.show()

# ===============================
# 🔥 4. HEATMAP (FULL COLOR)
# ===============================

plt.figure(figsize=(10, 6))
sns.heatmap(
    medical_df.corr(numeric_only=True),
    annot=True,
    cmap="magma",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.show()
```
