### The Gapminder provides population data from 1952 to 2007at 5 year intervals for several countries around the world .Compare the population of the european countries Frame United Kingdom Italy Germany and spain over this period using a like chart make appropriate modifications to the chart title ,axis,titles,legend,figure size,fontsize,colors etc to make the chart readable and visually appealling #Line Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px



df = pd.read_csv("D:\machinelearning\MachineLearning2026_PrinceParmar\Datasets\gapminderDataFiveYear.csv")

countries = ["United Kingdom", "Italy", "Germany", "Spain"]
eu_df = df[df["country"].isin(countries)]

plt.figure(figsize=(12, 7))

for country in countries:
    country_data = eu_df[eu_df["country"] == country]
    plt.plot(
        country_data["year"],
        country_data["pop"],
        marker='o',
        linewidth=2,
        label=country
    )

plt.title(
    "Population Growth of European Countries (1952â€“2007)",
    fontsize=16,
    fontweight="bold"
)
plt.xlabel("Year", fontsize=13)
plt.ylabel("Population", fontsize=13)
plt.legend(title="Country", fontsize=11)
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()

```
